```python
import pandas as pd 
import numpy as np 

```

```python
data_preprocessed = pd.read_csv('Absenteeism Preprocessed Dataset')
data_preprocessed.head()
```

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
data_preprocessed['Excessive Absenteeism'] = np.where(data_preprocessed['Absenteeism Time in Hours'] > 
                                                     data_preprocessed['Absenteeism Time in Hours'].median() , 1 , 0 ) 
```

```python
data_preprocessed
```

```python
data_preprocessed['Excessive Absenteeism'].value_counts()[1] 
```

```python
data_preprocessed['Excessive Absenteeism'].value_counts()[1] / data_preprocessed.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 
                                            'Daily Work Load Average', 
                                            'Distance to Work',
                                            'Day of the Week',
                                           # adding more columns through backward elimination 
                                           ],axis = 1 )
```

```python
data_with_targets
```

```python
# check if we have created a checkpoint or is data_with_target is referencing to the same object as data_preprocessed 

data_with_targets is data_preprocessed
```

```python
# => we have infact created a checkpoint 
```

```python
data_with_targets.iloc[:,:14]
```

```python
targets = data_with_targets['Excessive Absenteeism']
```

```python
targets
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1] 
unscaled_inputs
```

```python
unscaled_inputs.columns
```

```python
# col_to_scale = ['Month Value', 'Day of the Week',
#        'Transportation Expense', 'Distance to Work', 'Age',
#        'Daily Work Load Average', 'Body Mass Index', 'Children',
#        'Pets']

col_to_omit = ['Reason_1', 'Absence Reason_2', 'Absence Reason 3',
               'Absence Reason 4', 'Education']

col_to_scale = [ i for i in unscaled_inputs.columns.values if i not in col_to_omit ]
col_to_scale
# we are removing dummies from getting scaled as they will lose their interpretability of being dummies if they are scaled 
```

```python
from sklearn.preprocessing import StandardScaler 

absenteeism_scaler = StandardScaler()

unscaled_inputs_temp = unscaled_inputs

absenteeism_scaler.fit(unscaled_inputs_temp[col_to_scale])
```

```python
unscaled_inputs_temp[col_to_scale] = absenteeism_scaler.transform(unscaled_inputs_temp[col_to_scale]) 
scaled_inputs = unscaled_inputs_temp
```

```python

scaled_inputs
```

```python
scaled_inputs.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train , x_test , y_train , y_test = train_test_split(scaled_inputs, targets , train_size=0.8 , random_state=42)
```

```python
print(x_train.shape , y_train.shape)
```

```python
print(x_test.shape , y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import metrics 
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
# checking accuracy manually 
```

```python
model_output = reg.predict(x_train) 
```

```python
features = unscaled_inputs.columns
features
```

```python
summary_table = pd.DataFrame(columns = ['Features'] , data = features )
# summary_table['Coefficients'] = np.transpose(reg.coef_)
# another way to transpose 
summary_table['Coefficients'] = reg.coef_.T
```

```python
summary_table
```

```python
# we want to add intercept 
# but there is no method pre append 
# we can do it this way though 
```

```python
summary_table.index = summary_table.index + 1 
summary_table.loc[0] = ['Intercept' , reg.intercept_[0]]
```

```python
summary_table
```

```python
summary_table.sort_index()
```

```python
ordered_by_weights = summary_table.sort_values(by='Coefficients',ascending=False )
```

```python
ordered_by_weights
```

```python
ordered_by_weights['odd ratio'] = np.exp(ordered_by_weights.Coefficients)
```
